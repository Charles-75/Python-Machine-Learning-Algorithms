# Artificial Neural Networks

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

We want to build a neural network that predict if a bank customer will either remain in the bank or just leave according to certain features. 

```python
X = df.iloc[:,3:13].values
y = df.iloc[:,13].values

#Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
labelencoder_X_geography = LabelEncoder()
labelencoder_X_gender = LabelEncoder()
X[:,1] = labelencoder_X_geography.fit_transform(X[:,1])
X[:,2] = labelencoder_X_gender.fit_transform(X[:,2])

#onehotencoder_X_geography = OneHotEncoder(categorical_features=[1])
#X = onehotencoder_X_geography.fit_transform(X).toarray()
transformer = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), [1])], remainder='passthrough')
X = transformer.fit_transform(X.tolist())

X = X[:,1:] #avoiding the dummy trap

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

## Build our ANN

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense

#initialize the ANN
neural_network = Sequential()
#Add the input layer and the first hidden layer (with relu activation function)
neural_network.add(Dense(input_dim = 11, units = 6, kernel_initializer = 'uniform', activation = 'relu'))
#Add the second hidden layer
neural_network.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
#Add the output layer 
neural_network.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
#Compile the ANN
#We use the ADAM gradient descent optimization algorithm and the binary cross entropy as loss function (which is the same as the logistic regression loss function)
neural_network.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

neural_network.fit(X_train, y_train, batch_size = 64, epochs = 100)
```

## Evaluate our ANN

```python
y_pred = neural_network.predict(X_test)
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy : ',(cm[0][0]+cm[1][1])/y_test.shape[0])
```

## Better evaluate our ANN

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

def build_neural_network():
    neural_network = Sequential()
    neural_network.add(Dense(input_dim = 11, units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    neural_network.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    neural_network.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    neural_network.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return neural_network

neural_network = KerasClassifier(build_fn = build_neural_network, batch_size = 64, epochs = 100)
#n_jobs = -1 means all CPU are doing the computation
accuracies = cross_val_score(estimator = neural_network, X = X_train, y = y_train, cv = 10, n_jobs = -1)
print('Accuracies : ', accuracies)
print('Mean accuracy : ', accuracies.mean())
print('Standard deviation accuracy : ', accuracies.std())
```

## Improve our ANN  
We are going to use dropout technic in order to prevent our neural network to overfit.

```python
from keras.layers import Dropout

neural_network = Sequential()
neural_network.add(Dense(input_dim = 11, units = 6, kernel_initializer = 'uniform', activation = 'relu'))
neural_network.add(Dropout(rate = 0.1)) #20% chance for each unit of this layer to be desactivated for each iteration
neural_network.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu')) 
neural_network.add(Dropout(rate = 0.1))
neural_network.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
neural_network.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

neural_network.fit(X_train, y_train, batch_size = 64, epochs = 100)
```

```python
y_pred = neural_network.predict(X_test)
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy : ',(cm[0][0]+cm[1][1])/y_test.shape[0])
```

## Tune our ANN  
This gridsearch hyperparameters tuning will take very long time to run.

```python
from sklearn.model_selection import GridSearchCV

def build_neural_network(optimizer):
    neural_network = Sequential()
    neural_network.add(Dense(input_dim = 11, units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    neural_network.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    neural_network.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    neural_network.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return neural_network

neural_network = KerasClassifier(build_fn = build_neural_network)

parameters = {'batch_size': [25,32], 'epochs': [100, 500], 'optimizer': ['adam', 'rmsprop']}
grid_search = GridSearchCV(estimator = neural_network, param_grid = parameters, scoring = 'accuracy', cv = 10)
grid_search = grid_search.fit(X_train, y_train)

best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
